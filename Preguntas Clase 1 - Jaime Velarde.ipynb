## 3 Hemos construido un diccionario que se puede usar para encontrar el resultado si dos booleanos están conectados con el operador y

## Python no tiene un operador XOR, busque en Internet lo que se supone que debe hacer el operador XOR y cree un diccionario que proporcione el resultado de una operación XOR

```python
age = 40
blood_type = 'A'
my_and = {(True, True): True, (True, False): False, (False, True): False, (False, False): False }
print(my_and[age < 70, blood_type == 'B'])
```

```python
age = 40
blood_type = 'A'
def xor(age, blood_type):
    return bool((age and not blood_type) or (not age and blood_type))
print(xor(0,0))
print(xor(0,1))
print(xor(1,0))
print(xor(1,1))
```

## 5 Ejecute el siguiente programa, vea si puede encontrar algo extraño y explique el resultado usando su conocimiento de recolección de basura

```python
list_5 = [1, 2, 3]
print(f"The value of the list refered to by list_5 is {list_5} and its address is {id(list_5)}")
list_5 = [1 , 1, 3]
print(f"The value of the list refered to by list_5 is {list_5} and its address is {id(list_5)}")
list_5 = [1, 1, 4]
print(f"The value of the list refered to by list_5 is {list_5} and its address is {id(list_5)}")
```

Nota: Aunque en los tres casos hace referencia al nombre "lista_5", en cada caso, los valores que conforman la lista son distintos. La primera lista tiene el mismo ID que la tercera lista a pesar de no contener los mismos elementos. Esto se debe a que a Python esta programado de tal forma que recicla las variables, ocasionando el cambio de id.

## 6 Las cadenas son inmutables, pero ¿por qué funciona lo siguiente? ¿Claramente cambiamos el valor de la cadena llamada string_1? ¿Qué afirmaciones agregaría para aclarar lo que ha sucedido? usar id

```python
string_1 = "text 1"
print(string_1)
string_1 = "text 2"
print(string_1)
```

```python
string_1 = "text 1"
print(f"El valor de la cadena es {string_1} and its address is {id(string_1)}")
string_1 = "text 2"
print(f"El valor de la cadena es {string_1} and its address is {id(string_1)}")
```

Nota: El nombre del objeto es el mismo pero no contienen el mismo ID y sus elementos son diferentes, su id, indica que son cadenas distintas.
